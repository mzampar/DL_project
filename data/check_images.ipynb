```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pandas as pd
```

```python
id_df_rain = pd.read_csv("id_df_SRI_jpg_radar_1.csv")

id_df_rain['date_time'] = pd.to_datetime(id_df_rain['date_time'], format='%Y-%m-%d %H:%M:%S')

id_df_rain['time_diff'] = id_df_rain.iloc[::-1]['date_time'].diff()

print(len(id_df_rain[id_df_rain['time_diff'] < pd.Timedelta(minutes=11)]))
```

```python
def check_image(row):
    # Load image with Pillow
    image_path = f"../../fast/images/{row.id}.jpg"
    image = Image.open(image_path) # Convert to RGB if necessary

    # Convert image to numpy array
    pixels = np.array(image)

    # Reshape to list of RGB values
    pixels = pixels.reshape(-1)

    # Find unique colors and their counts
    unique_colors, counts = np.unique(pixels, axis=0, return_counts=True)

    # remove the white color
    unique_colors = unique_colors[1:]
    counts = counts[1:]

    # extract the max value for the counts
    max_value = np.max(counts)

    if row.name % 1000 == 0:
        print(row.name)
    
    return max_value
```

```python
def print_image(image_path):
    img = mpimg.imread(image_path)
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```

```python
id_df_rain['not_rain_pxl'] = id_df_rain.apply(check_image, axis=1)
id_df_rain.to_csv("id_df_rain.csv", index=False)
```
